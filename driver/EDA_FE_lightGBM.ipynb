#### 피처 요약표
- 피처 종류가 다양하니 한눈에 파악 가능하게 피처 요약표를 만들자. 
- 피처 요약표가 있으면 데이터 관리도 편하고, 추후 그래프를 그릴 때도 활용 가능

```python
# 피처 요약표 생성 함수
def resumetable(df):
    print(f'데이터 세트 형상: {df.shape}') # 데이터의 형태 출력
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입']) # 데이터 타입
    summary['결측값 개수'] = (df == -1).sum().values # 피처별 -1(결측값) 개수
    summary['고윳값 개수'] = df.nunique().values # 고윳값 개수
    summary['데이터 종류'] = None # 데이터 종류(이진, 명목형, 순서형)
    for col in df.columns: # for문을 순회하며 각 데이터 종류를 설정
        if 'bin' in col or col == 'target': # 해당 피처가 'bin' 데이터거나 타깃값이면 이진형 데이터
            summary.loc[col, '데이터 종류'] = '이진형'
        elif 'cat' in col: # 해당 피처가 'cat' 데이터라면 명목형 데이터
            summary.loc[col, '데이터 종류'] = '명목형'
        elif df[col].dtype == float: # 해당 피처값의 데이터 타입이 실수면 연속형 데이터
            summary.loc[col, '데이터 종류'] = '연속형'
        elif df[col].dtype == int: # 해당 피처값의 데이터 타입이 정수면 순서형 데이터
            summary.loc[col, '데이터 종류'] = '순서형'

    return summary
```

```python
summary = resumetable(train) # 훈련 데이터의 피처 요약표 저장
summary
```

## 1.2 데이터 시각화
- 타깃값 분포를 활용해 타깃값이 얼마나 불균형한지 알아보자. 
- 이진 피처, 명목형 피처, 순서형 피처의 고윳값별 타깃값 비율을 알아보자. 
- (막대그래프) 고윳값별 타깃값 비율을 보면 모델링 시 어떤 피처를 제거해야할지 알게 된다. 

### 1.2.1 타깃값 분포

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def write_percent(ax, total_size): # 도형 객체를 순회하며 막대 그래프 상단에 타깃값 비율 표시
    for patch in ax.patches:
        height = patch.get_height()     # 도형 높이(데이터 개수)
        width = patch.get_width()       # 도형 너비
        left_coord = patch.get_x()      # 도형 왼쪽 테두리의 x축 위치
        percent = height/total_size*100 # 타깃값 비율
        
        # (x, y) 좌표에 텍스트 입력
        ax.text(left_coord + width/2.0,     # x축 위치
                height + total_size*0.001,  # y축 위치
                '{:1.1f}%'.format(percent), # 입력 텍스트
                ha='center')                # 가운데 정렬
    
mpl.rc('font', size=15) # 폰트 사이즈 15
plt.figure(figsize=(7, 6)) # 그래프 사이즈

ax = sns.countplot(x='target', data=train) # 전체 훈련 데이터에 대한 타깃값 분포
write_percent(ax, len(train)) # 비율 표시
ax.set_title('Target Distribution'); # 그래프 제목
```

- 타깃값 0은 96.4%를 차지하며 1은 단 3.6%만을 차지한다. 
- 전체 운전자 중 3.6%만 보험금을 청구했다는 뜻
- 차 사고가 그리 흔하게 나오지 않으니 소수의 운전자만 보헙금을 청구했음을 알 수 있다. 
- 타깃값이 불균형하므로 비율이 작은 타깃값 1을 잘 예측하는 게 중요하다. 

##### 고윳값별 타깃값 1 비율
- 각 피처의 분포를 알아보기보다, 각 피처의 고윳값별 타깃값 1 비율을 알아보겠다. 
- 고윳값별 타깃값 1 비율을 통해 해당 피처가 모델링에 필요한 피처인지 확인할 수 있다. 

- 예컨대 피처 A에 고윳값 a, b가 있다고 하자. 
- 이때 고윳값 a, b별로 타깃값 비율이 얼마나 되는지 살펴보는 것이다. 
- 고윳값별로 타깃값 1 비율이 똑같거나 통계적 유효성이 떨어지면, 즉 통계적으로 유의미한 차이가 없다면 피처 A로는 무언가를 분별하기 어려우므로 예측에 도움이 되지 않는다. 
- 다시 말해, 고윳값에 따라 타깃값 비율이 달라야 그 피처가 타깃값 예측에 도움을 준다. 

##### 타깃값 1 비율의 통계적 유효성
- 타깃값 1 비율의 통계적 유효성이 떨어져도 불필요한 피처가 될 수 있다. 
- 통계적 유효성은 barplot()을 그릴 때 나타나는 신뢰구간으로 판단한다. 
- 신뢰구간이 좁다면 통계적으로 어느 정도 유효하다고 보고, 구간이 넓다면 신뢰하기 어렵다고 본다. 

##### 종합
- 고윳값별 타깃값 1 비율이 충분히 차이가 나고, 신뢰구간도 작은 피처여야 모델링에 도움이 된다. 
- 그렇지 않은 피처는 제거하는 것이 좋다. 

### 1.2.1 이진 피처

```python
def plot_target_ratio_by_features(data, features, target):
    """
    각 기능에 대한 대상 비율을 시각화합니다.

    Parameters:
    - data: DataFrame, 분석할 데이터
    - features: list, 대상 비율을 시각화할 기능의 목록
    - target: str, 대상 변수의 이름

    Returns:
    - None
    """
    for feature in features:
        plt.figure(figsize=(8, 6))
        sns.barplot(x=feature, y=target, data=data, ci=None)
        plt.title(f'{feature}에 따른 {target} 비율')
        plt.xlabel(feature)
        plt.ylabel(f'{target} 비율')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# 이진 피처들에서의 타깃값 1의 비율 시각화
bin_features = summary[summary['데이터 종류'] == '이진형'].index
plot_target_ratio_by_features(train, bin_features, 'target')
```

- 1번 그래프는 타깃값 그래프이므로 무시하고 2번부터 보면 된다. 
##### 2번의 ps_ind_06_bin 피처 그래프
- 값이 0일 때 타깃값이 1인 비율이 4% 정도다. 나머지 96%는 타깃값 0이다. 
- 값이 1일 때는 타깃값 1이 2.8% 정도 차지한다. 
- 종합: 고윳값별로 타깃값 비율이 다르다. 따라서 타깃값을 추정하는 예측력이 있다. 
- 신뢰구간도 좁기에 모델링에 도움이 되는 피처이다.

### 2.2. 평가지표 계산 함수 작성
- 본 경진대회의 평가지표는 정규화된 지니계수임
- 파이썬이나 사이킷런에서 기본 제공하지 않으므로 사용자 정의 함수를 직접 만들어야 한다. 
- 경제학에서 지니계수는 소득 불평등 정도를 나타내는 지표이다. 지니계수가 작을수록 소득 수준이 평등하고, 클수록 불평등함을 의미한다. 머신러닝에서 지니계수는 모델의 예측 성능을 측정하는 데 쓰인다. 예측값을 크기순으로 정렬해서 로렌츠 곡선을 구한다. 

#### 2.2.1. 정규화된 지니 계수 계산 함수
- 정규화란 값의 범위를 0~1 사이로 조정한다는 뜻이므로, 정규화 지니계수는 값이 0에 가까울수록 성능이 나쁘고, 1에 가까울수록 성능이 좋다는 의미가 된다. 
- 지금 맥락에서 정규화 지니계수는 '예측값에 대한 지니계수'를 '예측이 완벽할 때의 지니계수'로 나눈 값이다. 
- '예측값에 대한 지니계수'는 예측값과 실젯값으로 구한 지니계수를 뜻한다. 
- '예측이 완벽할 때의 지니계수'는 실젯값과 실제값으로 구한 지니계수를 뜻한다. 
- 예측이 완벽하다는 말은 예측값과 실젯값이 같다는 의미이기 때문

- 우리는 넘파이 내장함수를 이용해 구현할 것이다. 
- 실제 타깃값(y_true)와 예측 확률값(y_pred)을 입력받아 정규화 지니계수를 반환하도록 한다. 

```python
def eval_gini(y_true, y_pred): # 정규화된 지니 계수 계산 함수 구
    # 실제값과 예측값의 크기가 같은지 확인 (값이 다르면 오류 발생)
    assert y_true.shape == y_pred.shape

    n_samples = y_true.shape[0]                      # 데이터 개수
    L_mid = np.linspace(1 / n_samples, 1, n_samples) # 대각선 값

    # 1) 예측값에 대한 지니계수
    pred_order = y_true[y_pred.argsort()] # y_pred 크기순으로 y_true 값 정렬
    L_pred = np.cumsum(pred_order) / np.sum(pred_order) # 로렌츠 곡선
    G_pred = np.sum(L_mid - L_pred)       # 예측 값에 대한 지니계수

    # 2) 예측이 완벽할 때 지니계수
    true_order = y_true[y_true.argsort()] # y_true 크기순으로 y_true 값 정렬
    L_true = np.cumsum(true_order) / np.sum(true_order) # 로렌츠 곡선
    G_true = np.sum(L_mid - L_true)       # 예측이 완벽할 때 지니계수

    # 정규화된 지니계수
    return G_pred / G_true
```

```python
# 모델 훈련 시 검증 파라미터에 전달하기 위한 함수
# LightGBM용 gini() 함수
def gini(preds, dtrain):
    labels = dtrain.get_label()
    return 'gini', eval_gini(labels, preds), True # 반환값
```

### 2.3. 모델 훈련 및 성능 검증
#### 2.3.1. OOF 방식으로 LighGBM 훈련
- OOF 예측이란 K 폴드 교차검증을 수행하면서 각 폴드마다 테스트 데이터로 예측하는 방식이다. 
- K 폴드 교차검증을 하면서 폴드마다 1) 훈련 데이터로 모델을 훈련하고 2) 검증 데이터로 모델 성능을 측정하며 3) 테스트 데이터로 최종 타깃 확률도 예측한다. 
- 훈련된 모델로 마지막에 한 번만 예측하는 게 아니다. 
- 각 폴드별 모델로 여러 번 예측해 평균을 내는 방식이다. 

- 폴드마다 서로 다른 훈련 데이터와 검증 데이터를 이용해 모델을 훈련하고 평가한다. 
- 서로 다른 5개 모델을 만든 것
- 5개 모델이 각자 전체 테스트 데이터를 사용해 타깃 확률을 예측한다. 
- 그런 다음 예측 확률 5개의 평균을 구한다. 이 평균이 최종 예측 확률이다. 

- 먼저 층화 K 폴드 교차 검증기를 생성한다. 
- 본 경진대회의 타깃값은 불균형
- 따라서 K 폴드가 아니라 층화 K폴드를 수행하는 게 바람직
- 층화 K 폴드는 타깃값이 균등하게 배치되게 폴드를 나누는 방식임

```python
# 층화 K-폴드 교차 검증을 위한 StratifiedKFold 라이브러리 import
from sklearn.model_selection import StratifiedKFold

# StratifiedKFold 객체 생성
folds = StratifiedKFold(n_splits=5, shuffle=True, random_state=1991)
```

- StratifiedKFold는 n_splits 파라미터로 전달한 수만큼 폴드를 나눈다. 
- 여기서는 5개로 나눴다. 
- shuffle=True를 전달하면 폴드를 나눌 때 데이터를 섞어준다. 
- 훈련 데이터가 시계열 데이터가 아니라면 섞어주는 게 좋을 수도 있다. 
- 일부 폴드에만 특정 패턴 데이터가 몰려 있으면 모델 성능을 떨어뜨릴 수 있기 때문이다. 

- 이어서 LightGBM의 하이퍼파라미터를 설정한다. 
- LightGBM은 여러 하이퍼파라미터를 가지고 있지만, 여기서는 네 가지만 설정한다. 

```python
params = {'objective': 'binary',   # 이진분류
          'learning_rate': 0.01,   # 학습률 0.01
          'force_row_wise': True,  # 행별 히스토그램 빌딩 강제
          'random_state': 0}       # 반복시 같은 결과 도출
```

- 이진분류 문제라서 objective 파라미터를 binary로 설정한다. 
- 학습률은 0.01로, 랜덤 시드값은 0으로 설정했다. 
- force_row_wise=True는 경고 문구를 없애려고 추가한 파라미터이므로 신경쓰지 않아도 된다. 
- 다음으로 1차원 배열을 두 개 만들어준다.

```python
# OOF 방식으로 훈련된 모델로 검증 데이터 타깃값을 예측한 확률을 담을 1차원 배열
oof_val_preds = np.zeros(X.shape[0]) 
# OOF 방식으로 훈련된 모델로 테스트 데이터 타깃값을 예측한 확률을 담을 1차원 배열
oof_test_preds = np.zeros(X_test.shape[0]) 
```

- oof_val_preds는 검증 데이터를 활용해 예측한 확률값을 저장하는 배열이다. 
- K 폴드로 나눠도 훈련 데이터 전체가 결국엔 한 번씩 검증 데이터로 활용된다. 
- 따라서 oof_val_preds 배열 크기는 훈련 데이터와 같아야 한다. 
- 훈련 데이터 개수는 X.shape[0]로 구한다. 

- oof_test_preds는 테스트 데이터를 활용해 예측한 확률값을 저장하는 배열이다. 
- 최종 제출에 사용할 값이므로 크기는 테스트 데이터와 같아야 한다. 
- 테스트 데이터 개수는 X_test.shape[0]로 구한다.

- 둘다 초기 배열이므로 값은 0으로 채웠다. 
- np.zeros()는 넘파이 내장 함수로, 지정 개수만큼 0으로 채운 배열을 반환한다. 

- 이제 LightGBM 모델을 훈련한다. 
- 훈련하면서 OOF 예측도 수행할 것이다. (아래 코드는 OOF 예측 시 자주 사용하는 패턴이니 숙지!)

```python
import lightgbm as lgb
# 이전에 folds, X, y는 모두 이전 코드에서 정의한 변수들임
# 순서대로 층화 K 폴드 교차 검증기, 훈련데이터 피처, 훈련데이터의 타깃값임
# 아래 로직은 먼저 folds.split(X, y)를 호출해 데이터를 K 개로 나눠줌
# 파이썬 내장함수인 enumerate()를 적용해 인덱스(idx)와 K 폴드로 나뉜 훈련 데이터 인덱스(train_idx), 검증 데이터 인덱스(valid_idx)를 사용할 수 있게 한다. 
# 그리고 for문을 돌며 각 폴드마다 코드를 수행한다. 
for idx, (train_idx, valid_idx) in enumerate(folds.split(X, y)):
    print('#'*40, f'폴드 {idx+1} / 폴드 {folds.n_splits}', '#'*40)
    
    # train dataset 훈련용 데이터 나누기
    X_train, y_train = X[train_idx], y[train_idx]
    # validation dataset 검증용 데이터 나누기
    X_valid, y_valid = X[valid_idx], y[valid_idx]

    dtrain = lgb.Dataset(X_train, y_train) # LightGBM 전용 훈련 데이터셋으로 만들어 훈련
    dvalid = lgb.Dataset(X_valid, y_valid) # LightGBM 전용 검증 데이터셋으로 만들어 훈련

    params['verbose'] = 1000  # Move verbose_eval inside params

    lgb_model = lgb.train(params=params,
                          train_set=dtrain,
                          num_boost_round=10,
                          valid_sets=[dtrain, dvalid],
                          feval=gini) # feval 파라미터에는 gini를 전달, eval_gini는 정규화 지니계수 값을 계산하는 것인데, gini는 세 가지 값을 반환하는 함수로, 모델 훈련 시 사용하는 검증용 평가함수임
    
    # 훈련된 모델에 테스트 데이터를 주어 타깃 확률값 예측 : OOF 예측 방식
    oof_test_preds += lgb_model.predict(X_test)/folds.n_splits
    
    # 모델 성능 평가를 위해 검증 데이터를 활용해 타깃값을 예측
    # 맨 처음에는 oof_val_preds의 모든 값이 0으로 초기화되어있다. 
    # valid_idx에 해당되는 값만 검증 데이터 예측 확률로 업데이트해준다. 
    # 폴드가 5번 반복되면 oof_val_preds 내 모든 값이 검증 데이터 예측 확률로 업데이트된다. 
    # 최종 업데이트 된 oof_val_preds와 실제 타깃값 y를 비교해서 지니계수를 계산하면 모델 성능을 가늠 가능
    oof_val_preds[valid_idx] += lgb_model.predict(X_valid)
    
    # 현재 폴드에서 검증 데이터 예측 확률에 대한 정규화 지니계수를 계산
    # 현재 폴드에서 '검증 데이터 실제 타깃값'와 '예측 확률값'을 활용해 계산한 것
    # 폴드마다 계산되는 중간점검용 점수로 보면 된다. 
    gini_score = eval_gini(y_valid, oof_val_preds[valid_idx])
    print(f'폴드 {idx+1} 지니계수 : {gini_score}\n')
```

- 모델 훈련 시 verbose_eval =100으로 설정해서 100번째 이터레이션마다 성능 평가점수를 출력했다. 
- 성능 평가점수는 검증용 데이터로 계산한 값이다. 
- 로그 손실(logloss)은 이진분류할 때 lightGBM의 기본 평가 지표이다. 
- 이터레이션ㅇ르 반복할수록 지니계수가 커진다. = 모델 성능이 좋아진다. 
- 그러다 600번 이터레이션까지 결과를 출력하고 종료한다. 
- 모델 훈련 시 num_boost_rounds=1000을 전달했다. 
- 부스팅 최대 반복 횟수가 1,000번인데 다 채우지 못하고 조기종료되었다. 
- 조기종료 조건을 early_stopping_rounds=100로 지정해서, 100번 연속으로 지니계수가 최댓값을 갱신하지 못해 훈련을 멈춘 것이다. 
- 종료 시까지 정규화 지니계수가 가장 높았을 때는 583번째 이터레이션이며, 이때의 지니계수는 0.287804이다. 

```python
# 검증 데이터로 예측한 확률을 실제 타깃값과 비교해 지니계수를 출력한다. 
print('OOF 검증 데이터 지니계수:', eval_gini(y, oof_val_preds))
```

```python
import pandas as pd
import lightgbm as lgb
from sklearn.svm import SVC
from xgboost import XGBClassifier
from scipy import sparse
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from bayes_opt import BayesianOptimization
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
# 데이터 로드
train = pd.read_csv(data_path + 'train.csv', index_col='id')
test = pd.read_csv(data_path + 'test.csv', index_col='id')
submission = pd.read_csv(data_path + 'sample_submission.csv', index_col='id')
```

```python
# 전체 데이터 합치기
all_data = pd.concat([train, test], ignore_index=True)
all_data = all_data.drop('target', axis=1) # 타깃값 제거
```

```python
# 전체 피처
all_features = all_data.columns

# 명목형 피처
cat_features = [feature for feature in all_features if 'cat' in feature] 

# 원-핫 인코딩 적용
onehot_encoder = OneHotEncoder()
encoded_cat_matrix = onehot_encoder.fit_transform(all_data[cat_features]) 
```

```python
# '데이터 하나당 결측값 개수'를 파생 피처로 추가
all_data['num_missing'] = (all_data==-1).sum(axis=1)

# 연속형 피처만 선택
cont_features = [feature for feature in all_features if 'cat' not in feature and 'calc' not in feature] 
remaining_features = cont_features + ['num_missing']
```

```python
# ind 피처들을 합치기
ind_features = [feature for feature in all_features if 'ind' in feature]
all_data['mix_ind'] = all_data[ind_features].apply(lambda x: '_'.join(map(str, x)), axis=1)
```

```python
# mix_ind 피처로부터 카운트 피처 생성
all_data['mix_ind_count'] = all_data['mix_ind'].map(all_data['mix_ind'].value_counts())
```

```python
# 불필요한 피처 제거
drop_features = ['ps_ind_14', 'ps_ind_10_bin', 'ps_ind_11_bin', 
                 'ps_ind_12_bin', 'ps_ind_13_bin', 'ps_car_14']
all_data_remaining = all_data[remaining_features+['mix_ind_count']].drop(drop_features, axis=1)

# 범주형 피처와 연속형 피처를 sparse matrix로 변환
all_data_sprs = sparse.hstack([sparse.csr_matrix(all_data_remaining),
                               encoded_cat_matrix], format='csr')
```

```python
# 베이지안 최적화를 위한 하이퍼파라미터 범위 설정
param_bounds = {'num_leaves': (30, 40),
                'lambda_l1': (0.7, 0.9),
                'lambda_l2': (0.9, 1),
                'feature_fraction': (0.6, 0.7),
                'bagging_fraction': (0.6, 0.9),
                'min_child_samples': (6, 10),
                'min_child_weight': (10, 40)}

fixed_params = {'objective': 'binary',
                'learning_rate': 0.005,
                'bagging_freq': 1,
                'force_row_wise': True,
                'random_state': 1991}
```

```python
def eval_function_lgbm(num_leaves, lambda_l1, lambda_l2, feature_fraction,
                       bagging_fraction, min_child_samples, min_child_weight):
    '''LightGBM 모델의 평가지표(ROC AUC) 계산 함수'''
    params = {'num_leaves': int(round(num_leaves)),
              'lambda_l1': lambda_l1,
              'lambda_l2': lambda_l2,
              'feature_fraction': feature_fraction,
              'bagging_fraction': bagging_fraction,
              'min_child_samples': int(round(min_child_samples)),
              'min_child_weight': min_child_weight,
              'feature_pre_filter': False}
    params.update(fixed_params)
    
    # LightGBM 모델 훈련
    lgb_model = lgb.train(params=params, 
                           train_set=bayes_dtrain,
                           num_boost_round=2500,
                           valid_sets=bayes_dvalid)
    
    # 검증 데이터로 예측 수행
    preds = lgb_model.predict(X_valid)
    
    # ROC AUC 계산
    roc_auc = roc_auc_score(y_valid, preds)
    
    return roc_auc

def eval_function_svm(C, gamma):
    '''SVM 모델의 평가지표(ROC AUC) 계산 함수'''
    # SVM 모델 생성 및 훈련
    svm_model = SVC(C=C, gamma=gamma, probability=True)
    svm_model.fit(X_train, y_train)
    
    # 검증 데이터로 예측 수행
    preds = svm_model.predict_proba(X_valid)[:,1]
    
    # ROC AUC 계산
    roc_auc = roc_auc_score(y_valid, preds)
    
    return roc_auc

def eval_function_xgb(learning_rate, n_estimators, max_depth):
    '''XGBoost 모델의 평가지표(ROC AUC) 계산 함수'''
    # XGBoost 모델 생성 및 훈련
    xgb_model = XGBClassifier(learning_rate=learning_rate, 
                              n_estimators=int(round(n_estimators)), 
                              max_depth=int(round(max_depth)))
    xgb_model.fit(X_train, y_train)
    
    # 검증 데이터로 예측 수행
    preds = xgb_model.predict_proba(X_valid)[:,1]
    
    # ROC AUC 계산
    roc_auc = roc_auc_score(y_valid, preds)
    
    return roc_auc
```

```python
# 베이지안 최적화 객체 생성
optimizer_lgbm = BayesianOptimization(f=eval_function_lgbm, pbounds=param_bounds, random_state=0)
optimizer_svm = BayesianOptimization(f=eval_function_svm, pbounds={'C': (0.1), 'gamma': (1)}, random_state=0)
optimizer_xgb = BayesianOptimization(f=eval_function_xgb, pbounds={'learning_rate': (0.01), 'n_estimators': (100), 'max_depth': (10)}, random_state=0)
```

```python
# 베이지안 최적화 수행
optimizer_lgbm.maximize(init_points=3, n_iter=6)
optimizer_svm.maximize(init_points=3, n_iter=6)
optimizer_xgb.maximize(init_points=3, n_iter=6)
```

```python
# 최적의 하이퍼파라미터 추출
max_params_lgbm = optimizer_lgbm.max['params']
max_params_svm = optimizer_svm.max['params']
max_params_xgb = optimizer_xgb.max['params']
```

```python
# LightGBM, SVM, XGBoost 모델 생성
lgbm_model = lgb.LGBMClassifier(**{**max_params_lgbm, **fixed_params})
svm_model = SVC(probability=True, **max_params_svm)
xgb_model = XGBClassifier(**{**max_params_xgb})
```

```python
# 모델 훈련
lgbm_model.fit(X, y)
svm_model.fit(X, y)
xgb_model.fit(X, y)
```

```python
# 앙상블하여 예측
preds_lgbm = lgbm_model.predict_proba(X_test)[:,1]
preds_svm = svm_model.predict_proba(X_test)[:,1]
preds_xgb = xgb_model.predict_proba(X_test)[:,1]
preds_ensemble = (preds_lgbm + preds_svm + preds_xgb) / 3  # 앙상블 방법은 평균을 사용
```
