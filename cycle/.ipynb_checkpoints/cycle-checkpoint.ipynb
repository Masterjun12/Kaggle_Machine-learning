```python

train = pd.read_csv('/content/drive/MyDrive/bike-sharing-demand/train.csv')
test = pd.read_csv('/content/drive/MyDrive/bike-sharing-demand/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/bike-sharing-demand/sampleSubmission.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

selected_columns = train.columns[1:-1]

plt.figure(figsize=(12, 8))
sns.heatmap(train[selected_columns].corr(), annot=True)
plt.show()
```

```python
# 1. Remove outliers
train = train[train['weather'] != 4]
```

```python
# 2. concat data between train and test
data = pd.concat([train, test], ignore_index=True)
data.tail()
```

```python
# 3. add features
data['datetime'] = pd.to_datetime(data['datetime'])
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
data['hour'] = data['datetime'].dt.hour
data['weekday'] = data['datetime'].dt.weekday
```

```python
# 4. remove unrelated features
data.drop(['casual', 'registered', 'datetime', 'month', 'windspeed', 'season', 'atemp'], axis=1, inplace=True)
data.head()
```

```python
train.info()
```

```python
#5. split datasets
train = data[~pd.isnull(data['count'])]
test = data[pd.isnull(data['count'])]

X_train = train.drop(['count'], axis=1)
X_test = test.drop(['count'], axis=1)

y_train = train['count']
```

```python
# 유전 알고리즘 정의
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# Feature indices를 유전자로 설정
toolbox.register("attr_bool", np.random.choice, len(X_train.columns))
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=5)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
# 적합도 함수 정의
def evaluate(individual):
    X_train_sub, X_val_sub, y_train_sub, y_val_sub = train_test_split(X_train.iloc[:, individual], y_train, test_size=0.2, random_state=42)
    clf = RandomForestRegressor(n_estimators=100)
    clf.fit(X_train_sub, y_train_sub)
    y_pred = clf.predict(X_val_sub)
    return mean_squared_error(y_val_sub, y_pred),


toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
# 초기 모집단 생성
pop = toolbox.population(n=50)
```

```python
# 유전 알고리즘 실행
algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=10, verbose=True)
best_individual = tools.selBest(pop, k=1)[0]
```

```python
# 최적의 특성 선택 후 모델 학습
best_features = X_train.columns[best_individual]
final_model = RandomForestRegressor(n_estimators=100)
final_model.fit(X_train[best_features], y_train)
```

```python
# 테스트 데이터에 대한 예측 생성
test_predictions = final_model.predict(X_test[best_features])
```

```python
# 결과 CSV 파일 생성
submission_df = pd.DataFrame({'datetime': X_test.index, 'count': test_predictions})
submission_df['datetime'] = submission_df['datetime'].dt.strftime('%m/%d/%Y %H:%M')
submission_df.to_csv('sample_submission.csv', index=False)
```
