```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.ticker import MaxNLocator

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Set the style of matplotlib
%matplotlib inline
```

# 데이터 프로세싱

```python
# Load the training and testing datasets
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train
```

```python
# Extract features and target variable from training data
X_train = train.drop(columns=['id', 'FloodProbability'])
y_train = train['FloodProbability']
train
```

```python
# Extract features from test data
X_test = test.drop(columns=['id'])
test
```

### 타겟 시각화

```python
plt.style.use('fivethirtyeight')
# Visualization of target distribution
plt.figure(figsize=(16, 8))
plt.hist(train['FloodProbability'], bins=np.linspace(0.2825, 0.7275, 90),
         density=True, color='#E6E6FA', edgecolor='#006666')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('FloodProbability')
plt.ylabel('Density')
plt.show()
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **다층 퍼셉트론(Multi-Layer Perceptron)**

```python
# Initialize the MLP model
mlp_model = MLPRegressor(
    hidden_layer_sizes=(100,),
    activation='relu',
    solver='adam',
    alpha=0.0001,
    batch_size='auto',
    learning_rate='constant',
    learning_rate_init=0.001,
    power_t=0.5,
    max_iter=200,
    shuffle=True,
    random_state=42,
    tol=1e-4,
    verbose=True,
    warm_start=False,
    momentum=0.9,
    nesterovs_momentum=True,
    early_stopping=False,
    validation_fraction=0.1,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-8,
    n_iter_no_change=10,
    max_fun=15000
)

# Train the model
mlp_model.fit(X_train_scaled, y_train)

# Predict on the training set using the trained model
y_train_pred_mlp = mlp_model.predict(X_train_scaled)

mlp_train_mse = mean_squared_error(y_train, y_train_pred_mlp)
print(f'MLP Train MSE: {mlp_train_mse}')
```

```python
# Predict on the test data using MLP model
y_test_pred_mlp = mlp_model.predict(X_test_scaled)
```

```python

submission = pd.DataFrame({
    'id': test['id'],
    'FloodProbability': y_test_pred_mlp
})
submission
```

```python
submission.to_csv('submissionMLP.csv', index=False)
```
